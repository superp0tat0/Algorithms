```python
with open('data1.txt', 'r') as f:
    x = f.read().split("\n")

data_holder_diff = []
data_holder_div = []
for i in range(1, len(x)):
    current = list(map(int,x[i].split()))
    if(len(current) == 2): 
        data_holder_diff += [current + [current[0] - current[1]]]
        data_holder_div += [current + [current[0]/current[1]]]
```

```python
def sorting(data):
    if(len(data) <= 1):
        return data
    else:
        first_half = sorting(data[0:int(len(data)/2)])
        second_half = sorting(data[int(len(data)/2):len(data)])
        return merge_sort(first_half, second_half)

def merge_sort(arr1,arr2):
    result = []
    i = 0;j = 0
    while(i < len(arr1) and j < len(arr2)):
        if(arr1[i][2] > arr2[j][2]):
            result += [arr1[i]]
            i+=1
        elif(arr1[i][2] < arr2[j][2]):
            result += [arr2[j]]
            j+=1
        else:
            if(arr1[i][0] > arr2[j][0]):
                result += [arr1[i]]
                i+=1
            else:
                result += [arr2[j]]
                j+=1
    
    if(i == len(arr1)):
        result = result + arr2[j:len(arr2)]
    else:
        result = result + arr1[i:len(arr1)]
    return result

def weighted_complete_time(data):
    finish = 0
    result = 0
    
    for i in data:
        finish += i[1]
        result += finish*i[0]
    
    return result
```

```python
sorted_data_diff = sorting(data_holder_diff)
sorted_data_div = sorting(data_holder_div)
print(weighted_complete_time(sorted_data_diff))
print(weighted_complete_time(sorted_data_div))
```

```python
with open('data2.txt', 'r') as f:
    x = f.read().split("\n")

data_holder = []

for index in range(1,len(x)-1):
    data_split = x[index].split()
    data_holder.append((int(data_split[0]), int(data_split[1]), int(data_split[2])))
```

```python
def prim(data):
    all_vertex = set(list(range(1,501)))
    visited = set([3])
    total = 0
    
    while(visited != all_vertex):
        print(visited)
        edge = get_cheapest(data, visited)
        total += edge[2]
        visited.add(edge[1])
        visited.add(edge[0])
        
    return total

def get_cheapest(data, visited):
    non_visited = set(list(range(1,501))) - visited
    filtered = []
    for element in data:
        if((element[0] in visited and element[1] in non_visited) or 
           (element[1] in visited and element[0] in non_visited)):
            filtered.append(element)
    minimum = min(filtered,key = lambda x:(x[2]))
    return minimum
    
```

```python
print(prim(data_holder))
```
