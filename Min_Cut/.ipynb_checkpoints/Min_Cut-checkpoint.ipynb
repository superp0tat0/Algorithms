```python
import numpy as np
import copy

#File preprocessing
def remove_blank(x):
    pre_blank = 0
    af_blank = 0
    
    for i in range(0, len(x)):
        if(x[i] != " "):
            pre_blank = i
            break
    
    for i in range(len(x)-1, 0, -1):
        if(x[i] != " "):
            af_blank = i
            break
    
    return x[pre_blank:af_blank+1]
```

```python
def min_cut(dataset):
    if(len(dataset) == 2):
        return dataset
    else:
        #Randomly select two entry to begin, do not want duplicates
        value1 = 0; value2 = 0
        while(value1 == value2):
            value1 = np.random.randint(low = 0, high = len(dataset))
            value2 = np.random.randint(low = 0, high = len(dataset))
        
        #merge two node to one
        new_data = merge(dataset, value1, value2)
        return min_cut(new_data)

def merge(dataset, value1, value2):
    #merge the edges, get the index of two node first
    index1 = dataset[value1][0]
    index2 = dataset[value2][0]
    new_edge = dataset[value1] + dataset[value2]
    
    #remove self loop edges
    i = 0
    while(i <= len(new_edge)):
        if(new_edge[i] == index1 or new_edge[i] == index2):
            new_edge.pop(i)
        else:
            i+=1
    
    #update the new edge to value1 and index1, remove the old edge of value2
    dataset[value1] = [index1] + new_edge
    dataset.pop(value2)
    
    #update all the connections, the connect to index 2 should be index1 now
    for i in range(0, len(dataset)):
        for j in range(0, len(dataset[i])):
            if(dataset[i][j] == index2):
                dataset[i][j] = index1
    
    return dataset

```

```python
f = open("Program_Input.txt", "r")
data = []
counter = 0
for i in f:
    current = remove_blank(i.replace("\n","").replace("\t"," ")).split(" ")
    data.append(list(map(int,current)))

print(data)
```

```python
length = []
for sed in range(0,3000):
    # make a copy of the current data
    copy_data = copy.deepcopy(data)
    
    result = min_cut(copy_data)
    length.append(len(result[0]))
    length.append(len(result[1]))

print(min(length))
print(length)
```
