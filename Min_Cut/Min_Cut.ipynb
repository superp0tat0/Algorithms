```python
import numpy as np
import random
import copy

#File preprocessing
def remove_blank(x):
    pre_blank = 0
    af_blank = 0
    
    for i in range(0, len(x)):
        if(x[i] != " "):
            pre_blank = i
            break
    
    for i in range(len(x)-1, 0, -1):
        if(x[i] != " "):
            af_blank = i
            break
    
    return x[pre_blank:af_blank+1]
```

```python
def clean_list(dataset):
    counter = 0
    while(counter < len(dataset)):
        if not dataset[counter]:
            dataset.pop(counter)
        else:
            counter += 1
    
    return dataset
            
def min_cut(dataset, avaliable_list):
    if(len(avaliable_list) == 2):
        return clean_list(dataset)
    else:
        #Randomly select two entry to begin, do not want duplicates
        index1 = random.choice(avaliable_list)
        index2 = random.choice(dataset[index1])-1
        
        #merge two node to one
        
        Error = True
        while(Error):
            try:
                avaliable_list.remove(index2)
                Error = False
            except:
                index2 = random.choice(dataset[index1])-1
    
        new_data = merge(dataset, index1, index2)
        
        return min_cut(new_data, avaliable_list)

def merge(dataset, index1, index2):
    #merge the edges, get the index
    new_edge = dataset[index1] + dataset[index2]
    
    #remove self loop edges
    counter = 0
    while(counter < len(new_edge)):
        if(new_edge[counter] == index1+1 or new_edge[counter] == index2+1):
            new_edge.pop(counter)
        else:
            counter += 1
    
    #update the new edge to index1, remove the old edge of index2
    dataset[index1] = new_edge
    dataset[index2] = [] #Disconnect all
    
    #update all the connections, the connect to index 2 should be index1 now
    for i in range(0, len(dataset)):
        for j in range(0, len(dataset[i])):
            if(dataset[i][j] == index2+1):
                dataset[i][j] = index1+1
    
    return dataset
```

```python
f = open("Program_Input.txt", "r")
data = []
counter = 0
for i in f:
    current = remove_blank(i.replace("\n","").replace("\t"," ")).split(" ")
    data.append(list(map(int,current)))
```

```python
length = []
for sed in range(0,1000):
    # make a copy of the current data
    copy_data = copy.deepcopy(data)
    ava_list = list(range(0,200))
    
    result = min_cut(copy_data,ava_list)
    length.append(len(result[0]))
    length.append(len(result[1]))

print(min(length))
print(length)
```
