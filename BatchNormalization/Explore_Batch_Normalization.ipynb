```python
import numpy as np
import random as random
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
```

```python
size = 10
mu = 0
sig = 1

x_train = np.linspace(-size*np.pi, size*np.pi, 100*size)
noise = np.random.normal(loc=mu, scale=sig, size=100*size)
y_real = 5*np.sin(x_train)
y_train = 5*np.sin(x_train) + noise
y_train_scale = (y_train - np.mean(y_train))/np.std(y_train)
```

```python
plt.plot(x_train, y_train, 'b')
plt.plot(x_train, y_train_scale, 'r')
plt.plot(x_train, y_real, 'g')
plt.show
```
