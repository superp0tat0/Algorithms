```python
import numpy as np
import pandas as pd
import random as random
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0, 9.0)
```

```python
def cal(X):
    Y = []
    for i in X:
        Y.append(3.8*i + 5.1 + np.random.normal(0,10,1)[0])
    return Y
```

```python
def com_plot(Y_pred):
    plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # regression line
    plt.show()
```

```python
X = [random.uniform(0,30) for i in range(60)]
Y = cal(X)
X = pd.Series(X)
Y = pd.Series(Y)
```

```python
#Show the plot
plt.scatter(X, Y)
plt.show()
```

```python
#the predict function param
w = 0; b = 0
#the learning parameter:
learn = 0.001; epoch = 1000; n = len(X)
loss = []
```

```python
# Performing Gradient Descent 
plt.scatter(X, Y) 
for i in range(epoch): 
    Y_pred = w*X + b  # The current predicted value of Y
    D_w = (-2/n) * sum(X * (Y - Y_pred))
    D_b = (-2/n) * sum(Y - Y_pred)
    w = w - learn * D_w
    b = b - learn * D_b
    
    #Plot
    com_plot(Y_pred)
    plt.cla()
    loss.append((1/n) * sum((Y-Y_pred) * (Y-Y_pred)))
```
